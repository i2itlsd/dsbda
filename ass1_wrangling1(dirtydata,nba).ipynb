```python
df['Calories'] = df['Calories'].abs()
print(df)
```

```python
x = df['Calories'].mean()
x
```

```python
# Filling null values
df['Calories'].fillna(x, inplace=True)
print(df)
```

```python
# Convert the date into a "Date" format
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
print(df)
```

```python
# 7th location's duration value is 450 which is an outlier
df.loc[7, 'Duration'] = 45
print(df)
```

```python
# Finding duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
print(df)
```
